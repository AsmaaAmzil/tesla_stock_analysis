# 2. Feature Engineering for Tesla Stock Prediction

This notebook focuses on creating and preparing features for our prediction model.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Load the preprocessed data
df = pd.read_csv("../data/TSLA.csv")
df['Date'] = pd.to_datetime(df['Date'])
```

## 2.1 Creating Technical Indicators

```python
def add_technical_indicators(df):
    # Moving averages
    df['MA5'] = df['Close'].rolling(window=5).mean()
    df['MA20'] = df['Close'].rolling(window=20).mean()
    
    # Daily returns
    df['Daily_Return'] = df['Close'].pct_change()
    
    # Price range
    df['Daily_Range'] = df['High'] - df['Low']
    
    # Volume features
    df['Volume_MA5'] = df['Volume'].rolling(window=5).mean()
    df['Volume_Ratio'] = df['Volume'] / df['Volume_MA5']
    
    # Volatility (standard deviation of returns)
    df['Volatility'] = df['Daily_Return'].rolling(window=20).std()
    
    # Price momentum
    df['Momentum'] = df['Close'] - df['Close'].shift(5)
    
    return df

df = add_technical_indicators(df)
print("Added technical indicators:")
print(df.columns.tolist())
df.info()
```

## 2.2 Feature Selection and Preprocessing

```python
# Remove rows with NaN values
df_clean = df.dropna()

# Select features for modeling
feature_columns = [
    'Open', 'High', 'Low', 'Volume',
    'MA5', 'MA20', 'Daily_Return', 'Daily_Range',
    'Volume_Ratio', 'Volatility', 'Momentum'
]

# Target variable
target = 'Close'

# Scale the features
scaler = StandardScaler()
df_scaled = pd.DataFrame(
    scaler.fit_transform(df_clean[feature_columns]),
    columns=feature_columns,
    index=df_clean.index
)

# Add target variable back
df_scaled[target] = df_clean[target]
```

## 2.3 Feature Importance Analysis

```python
# Calculate correlation with target
correlations = df_scaled[feature_columns].corrwith(df_scaled[target])

# Plot feature correlations
plt.figure(figsize=(12, 6))
correlations.sort_values().plot(kind='bar')
plt.title('Feature Correlations with Close Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 2.4 Save Processed Data

```python
# Save processed dataset
df_scaled.to_csv('../data/TSLA_processed.csv', index=False)
print("Saved processed dataset with shape:", df_scaled.shape)
```
