# 1. Tesla Stock Data Exploration

This notebook performs exploratory data analysis on Tesla (TSLA) stock data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Enable inline plotting in Jupyter Notebook
%matplotlib inline

# Apply Seaborn theme
sns.set_theme(style="whitegrid")
```

## 1.1 Data Loading and Initial Inspection

```python
# Load the data
df = pd.read_csv("../data/TSLA.csv")

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Display basic information about the dataset
print("Dataset Info:")
print(df.info())

print("\nFirst few rows:")
print(df.head())

print("\nBasic statistics:")
print(df.describe())
```

## 1.2 Time Series Analysis

```python
# Plot closing prices over time
#plt.figure(figsize=(15, 6))
plt.figure(figsize=(5, 3))
plt.plot(df['Date'], df['Close'])
plt.title('Tesla Stock Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Observations:

 We can observe several distinct phases in the company's valuation. The stock maintained relatively stable low prices under $100 from 2011 through early 2013, followed by a significant surge to around $200-300 range in 2014. From 2014 to early 2019, the price showed considerable volatility but generally traded in a channel between $150-400. The most dramatic feature is the extraordinary upward trajectory in late 2019 into 2020, where the stock price skyrocketed to nearly $800, representing an unprecedented increase in market valuation. 

## 1.3 Correlation Analysis

```python
# Calculate correlation matrix
correlation_matrix = df[['Open', 'High', 'Low', 'Close', 'Volume']].corr()

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of TSLA Stock Features')
plt.tight_layout()
plt.show()
```

This correlation matrix reveals interesting relationships between Tesla stock's price metrics (Open, High, Low, Close) and trading volume. The price metrics show perfect correlation (1.0) with each other, which is expected as they naturally move together during trading. However, what's particularly noteworthy is the moderate positive correlation (approximately 0.5) between trading volume and all price metrics. This suggests that periods of higher trading activity tend to coincide with higher prices, though the relationship isn't strongly deterministic. The uniformity of the volume correlations (all around 0.5) indicates that volume's relationship with price remains consistent regardless of which price point (open, high, low, or close) is considered.

## 1.4 Volatility Analysis

```python
# Calculate daily price range (volatility)
df['Daily_Range'] = df['High'] - df['Low']

# Plot volatility over time
plt.figure(figsize=(15, 6))
plt.plot(df['Date'], df['Daily_Range'])
plt.title('Daily Price Range (Volatility) Over Time')
plt.xlabel('Date')
plt.ylabel('Price Range ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 1.5 Volume Analysis

```python
#Plot volume over time
plt.figure(figsize=(15, 6))
plt.bar(df['Date'], df['Volume'], alpha=0.8, color='darkblue', edgecolor='black')
plt.title('Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 1.6 Statistical Analysis

```python
# Calculate daily returns
df['Daily_Return'] = df['Close'].pct_change() * 100

# Plot distribution of daily returns
plt.figure(figsize=(10, 6))
sns.histplot(df['Daily_Return'].dropna(), kde=True)
plt.title('Distribution of Daily Returns')
plt.xlabel('Daily Return (%)')
plt.ylabel('Frequency')
plt.show()

# Print summary statistics of daily returns
print("\nDaily Returns Statistics:")
print(df['Daily_Return'].describe())
```
